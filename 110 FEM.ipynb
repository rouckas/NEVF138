```python
from dolfin import *
from IPython.display import Image
```

```python
mesh = UnitSquareMesh(6, 6)
```

```python
wiz = plot(mesh)
wiz.write_png("mesh")
Image("mesh.png")
```

```python
V = FunctionSpace(mesh, 'Lagrange', 1)
```

```python
u0 = Expression('1 + x[0]*x[0] + 2*x[1]*x[1]')
```

```python
def u0_boundary(x, on_boundary):
    return on_boundary
```

```python
bc = DirichletBC(V, u0, u0_boundary)
```

```python
u = TrialFunction(V)
v = TestFunction(V)
```

```python
f = Constant(50.0)
a = inner(nabla_grad(u), nabla_grad(v))*dx
L = f*v*dx
```

```python
# Compute solution
u = Function(V)
solve(a == L, u, bc)
```

```python
# Plot solution and mesh
wiz = plot(u)
wiz.write_png("u")
Image("u.png")
```
