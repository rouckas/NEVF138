```python
%matplotlib inline
from pylab import *
from numpy import *
```

```python
x = linspace(-1., 1., 400)
y = linspace(-1., 1., 400)
X, Y = meshgrid(x, y)
```

```python
X = X + 1j*Y
```

```python
# http://nbviewer.ipython.org/github/empet/Math/blob/master/DomainColoring.ipynb
figure(figsize=(12,6))
subplot(121)
def Hcomplex(z):# computes the hue corresponding to the complex number z
    H=angle(z)/(2*pi)+1
    return mod(H,1)
def g(x):
    return (1.0-1/(1+x**2))**0.2

from matplotlib.colors import hsv_to_rgb
H = Hcomplex(X)
V = log(abs(X))
V1 = (V - floor(V))
V2 = ((angle(X)/(2*pi)+0.5)*2).clip(0,1)
V3 = ((angle(X)/(2*pi)+.5)*2).clip(0,1)

V=(V1)
S = ones_like(H, float)
H = ones_like(H, float)

HSV = dstack((H,S*0,V))# V**0.2>V for V in[0,1];this choice  avoids too dark colors
RGB=hsv_to_rgb(HSV)
HSV2 = dstack((H,S,(1-V1)*(1-V3)))# V**0.2>V for V in[0,1];this choice  avoids too dark colors
HSV2 = dstack((H,S,V3))# V**0.2>V for V in[0,1];this choice  avoids too dark colors
RGB2=hsv_to_rgb(HSV2)


imshow(RGB)
subplot(122)
imshow(RGB2)
```
